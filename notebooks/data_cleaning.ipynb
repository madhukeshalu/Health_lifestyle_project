```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip()
```

```python
print("After deleting person_id nulls:", df.shape)
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['age'].fillna(df['age'].median(), inplace=True)
```

```python
categorical_cols = [
    'gender','diet_quality','stress_level',
    'smoking_status','alcohol_consumption',
    'mental_wellbeing_score',
    'chronic_conditions','family_history'
]

for col in categorical_cols :
    df[col] = df[col].astype(str).str.strip().str.lower()
```

```python
num_cols = [
    'age', 'height_cm', 'weight_kg',
    'physical_activity_hours_per_week',
    'sleep_hours_per_day',
    'mental_wellbeing_score'
]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df.fillna({col: df[col].median()}, inplace=True)
```

```python
# Numeric columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Categorical columns (except person_id)
categorical_cols = df.select_dtypes(include='object').columns
categorical_cols = categorical_cols.drop('person_id', errors='ignore')

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df = df.drop_duplicates(subset='person_id')
```

```python
import numpy as np

# ---------- Gender ----------
df['gender'] = df['gender'].astype(str).str.lower().str.strip()
df['gender'] = df['gender'].fillna(df['gender'].mode()[0])
df['gender'] = df['gender'].map({'male': 1, 'female': 0})

# ---------- Diet Quality (ordinal) ----------
df['diet_quality'] = df['diet_quality'].astype(str).str.lower().str.strip()
df['diet_quality'] = df['diet_quality'].fillna('average')

diet_map = {
    'poor': 1,
    'average': 2,
    'good': 3
}
df['diet_quality'] = df['diet_quality'].map(diet_map)

# ---------- Stress Level (ordinal) ----------
df['stress_level'] = df['stress_level'].astype(str).str.lower().str.strip()

stress_map = {
    'low': 1,
    'medium': 2,
    'high': 3
}
df['stress_level'] = df['stress_level'].map(stress_map)

# ---------- Smoking Status ----------
df['smoking_status'] = df['smoking_status'].astype(str).str.lower().str.strip()

smoke_map = {
    'non-smoker': 0,
    'occasional': 1,
    'regular': 2
}
df['smoking_status'] = df['smoking_status'].map(smoke_map)

# ---------- Alcohol Consumption ----------
df['alcohol_consumption'] = df['alcohol_consumption'].astype(str).str.lower().str.strip()
df['alcohol_consumption'] = df['alcohol_consumption'].fillna('moderate')

alcohol_map = {
    'moderate': 1,
    'high': 2
}
df['alcohol_consumption'] = df['alcohol_consumption'].map(alcohol_map)

# ---------- Family History ----------
df['family_history'] = df['family_history'].astype(str).str.lower().str.strip()
df['family_history'] = df['family_history'].map({'yes': 1, 'no': 0})

# ---------- Chronic Conditions (binary) ----------
df['chronic_conditions'] = (
    df['chronic_conditions']
    .replace(['nan', 'NaN', np.nan], 'none')
    .astype(str)
    .str.lower()
    .str.strip()
)

df['chronic_conditions'] = df['chronic_conditions'].apply(
    lambda x: 0 if x == 'none' else 1
)

```

```python
df['mental_wellbeing_score'] = pd.to_numeric(
    df['mental_wellbeing_score'], errors='coerce'
)
df['mental_wellbeing_score'] = df['mental_wellbeing_score'].fillna(
    df['mental_wellbeing_score'].median()
)

```

```python
num_cols = [
    'height_cm','weight_kg',
    'physical_activity_hours_per_week',
    'sleep_hours_per_day',
    'daily_screen_time_hours',
    'water_intake_liters',
    'fast_food_frequency_per_week'
]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())



```

```python
df = df[df['age'].between(5, 100)]
df = df[df['sleep_hours_per_day'].between(3, 14)]
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['gender', 'alcohol_consumption'])
```

```python
import os

base_path = "../data/cleaned"
os.makedirs(base_path, exist_ok=True)

df.to_csv(f"{base_path}/health_lifestyle_cleaned.csv", index=False)


```
