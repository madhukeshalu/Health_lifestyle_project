```python
import pandas as pd
import numpy as np

# Load cleaned dataset
file_path = r"C:\Users\Madhukeshalu\Documents\Health_lifestyle_project\data\cleaned\health_lifestyle_cleaned.csv"
df = pd.read_csv(file_path)

df.head()

```

```python
required_cols = [
    'height_cm',
    'weight_kg',
    'sleep_hours_per_day',
    'physical_activity_hours_per_week',
    'stress_level',
    'mental_wellbeing_score',
    'diet_quality'
]

missing = [col for col in required_cols if col not in df.columns]

if missing:
    print("Missing columns:", missing)
else:
    print("All required columns are present ")
```

```python
df['bmi'] = df['weight_kg'] / (df['height_cm'] ** 2)
```

```python
df['wellness_score'] = (
    df['physical_activity_hours_per_week'] +
    df['sleep_hours_per_day'] +
    df['mental_wellbeing_score'] +
    (df['diet_quality'] * 2) -
    (df['stress_level'] * 2)
) / 5
```

```python
def get_risk_category(row):
    if row['wellness_score'] < 4 or row['stress_level'] > 7:
        return 'High Risk'
    elif row['wellness_score'] < 6:
        return 'Medium Risk'
    else:
        return 'Low Risk'

df['risk_category'] = df.apply(get_risk_category, axis=1)
```

```python
status_map = {
    'Low': 'Healthy',
    'Medium': 'Needs Attention',
    'High': 'At Risk'
}

df['health_status'] = df['risk_category'].map(status_map)
```

```python
def get_health_status(score):
    if score < 4:
        return 'Poor'
    elif score < 7:
        return 'Average'
    else:
        return 'Good'

df['health_status'] = df['wellness_score'].apply(get_health_status)
```

```python
def health_suggestion(risk):
    if risk == 'Low':
        return "Maintain your healthy lifestyle."
    elif risk == 'Medium':
        return "Increase physical activity, improve diet, and manage stress."
    else:
        return "High health risk detected. Immediate lifestyle changes and medical consultation recommended."

df['health_suggestion'] = df['risk_category'].apply(health_suggestion)
```

```python
def get_risk_reason(row):
    reasons = []

    if row['stress_level'] > 7:
        reasons.append("High Stress")
    if row['sleep_hours_per_day'] < 6:
        reasons.append("Poor Sleep")
    if row['physical_activity_hours_per_week'] < 2:
        reasons.append("Low Physical Activity")
    if row['bmi'] >= 30:
        reasons.append("Obesity")
    elif row['bmi'] < 18.5:
        reasons.append("Underweight")

    return ", ".join(reasons) if reasons else "No Major Risk"

df['risk_reason'] = df.apply(get_risk_reason, axis=1)
```

```python
def get_priority_focus(row):
    if row['stress_level'] > 7:
        return "Stress Management"
    elif row['sleep_hours_per_day'] < 6:
        return "Improve Sleep"
    elif row['physical_activity_hours_per_week'] < 2:
        return "Increase Physical Activity"
    elif row['diet_quality'] < 5:
        return "Improve Diet"
    else:
        return "Maintain Healthy Lifestyle"

df['priority_focus'] = df.apply(get_priority_focus, axis=1)
```

```python
def get_health_suggestion(row):
    if row['risk_category'] == 'High Risk':
        return "Consult healthcare professional and adopt immediate lifestyle changes"
    elif row['risk_category'] == 'Medium Risk':
        return "Improve sleep, reduce stress, and increase daily physical activity"
    else:
        return "Maintain current healthy habits and do regular health checkups"

df['health_suggestion'] = df.apply(get_health_suggestion, axis=1)
```
